```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 定义 Prewitt 算子
prewitt_x = np.array([[-1, 0, 1], 
                      [-1, 0, 1], 
                      [-1, 0, 1]], dtype=np.float32)

prewitt_y = np.array([[-1, -1, -1], 
                      [0, 0, 0], 
                      [1, 1, 1]], dtype=np.float32)

# 对图像应用 Prewitt 算子
edges_x = cv2.filter2D(image, cv2.CV_64F, prewitt_x)
edges_y = cv2.filter2D(image, cv2.CV_64F, prewitt_y)

# 计算梯度幅值和方向
gradient_magnitude = cv2.magnitude(edges_x, edges_y)
gradient_direction = cv2.phase(edges_x, edges_y, angleInDegrees=True)

# 归一化梯度幅值
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 1, cv2.NORM_MINMAX)

# 使用 Matplotlib 显示结果
plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gradient Magnitude')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 读取图像
img = cv2.imread('D:\\picture\\school.png')
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 灰度化处理图像
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Prewitt算子
kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]], dtype=int)
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=int)

# 使用 Prewitt 算子计算 x 和 y 方向的梯度
x = cv2.filter2D(grayImage, cv2.CV_16S, kernelx)
y = cv2.filter2D(grayImage, cv2.CV_16S, kernely)

# 转换为 uint8
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)

# 合成最终的 Prewitt 边缘图像
Prewitt = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)

# 应用阈值化，使边缘更加清晰
_, thresholded = cv2.threshold(Prewitt, 50, 255, cv2.THRESH_BINARY)  # 50为阈值，可以根据需要调整

# 用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']

# 使用 Matplotlib 显示结果
plt.figure(figsize=(14, 10))

# 显示原始图像
plt.subplot(131), plt.imshow(grayImage, cmap=plt.cm.gray), plt.title('原始图像'),plt.axis('off')

# 显示 Prewitt 边缘图像
plt.subplot(132), plt.imshow(Prewitt, cmap=plt.cm.gray), plt.title('Prewitt算子'), plt.axis('off')

# 显示阈值化后的边缘图像
plt.subplot(133), plt.imshow(thresholded, cmap=plt.cm.gray), plt.title('阈值化后的边缘'), plt.axis('off')
plt.show()
```
