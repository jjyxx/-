```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

#  Prewitt 
prewitt_x = np.array([[-1, 0, 1], 
                      [-1, 0, 1], 
                      [-1, 0, 1]], dtype=np.float32)

prewitt_y = np.array([[-1, -1, -1], 
                      [0, 0, 0], 
                      [1, 1, 1]], dtype=np.float32)

#  Prewitt 
edges_x = cv2.filter2D(image, cv2.CV_64F, prewitt_x)
edges_y = cv2.filter2D(image, cv2.CV_64F, prewitt_y)

# 
gradient_magnitude = cv2.magnitude(edges_x, edges_y)
gradient_direction = cv2.phase(edges_x, edges_y, angleInDegrees=True)

# 
gradient_magnitude = cv2.normalize(gradient_magnitude, None, 0, 1, cv2.NORM_MINMAX)

#  Matplotlib 
plt.figure(figsize=(12, 8))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Gradient Magnitude')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 
img = cv2.imread('D:\\picture\\school.png')
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Prewitt
kernelx = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]], dtype=int)
kernely = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=int)

#  Prewitt  x  y 
x = cv2.filter2D(grayImage, cv2.CV_16S, kernelx)
y = cv2.filter2D(grayImage, cv2.CV_16S, kernely)

#  uint8
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)

#  Prewitt 
Prewitt = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)

# 
_, thresholded = cv2.threshold(Prewitt, 50, 255, cv2.THRESH_BINARY)  # 50

# 
plt.rcParams['font.sans-serif'] = ['SimHei']

#  Matplotlib 
plt.figure(figsize=(14, 10))

# 
plt.subplot(131), plt.imshow(grayImage, cmap=plt.cm.gray), plt.title(''),plt.axis('off')

#  Prewitt 
plt.subplot(132), plt.imshow(Prewitt, cmap=plt.cm.gray), plt.title('Prewitt'), plt.axis('off')

# 
plt.subplot(133), plt.imshow(thresholded, cmap=plt.cm.gray), plt.title(''), plt.axis('off')
plt.show()
```
