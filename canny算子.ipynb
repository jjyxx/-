```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
src = cv2.imread("D:/picture/school.png")
if src is None:
    print("Could not load image...")
    exit()

# 转换为灰度图像
gray_src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# 使用 Canny 边缘检测
# Canny 算子需要两个阈值，低阈值和高阈值
low_threshold = 100
high_threshold = 200
edges = cv2.Canny(gray_src, low_threshold, high_threshold)

# 使用 plt 显示图像
plt.figure(figsize=(10, 8))
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 原始图像
plt.subplot(1, 2, 1)
#plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))  # 转换为 RGB 格式以适应 matplotlib
plt.imshow(gray_src,cmap='gray')
plt.title("原始图像")
plt.axis('off')  # 不显示坐标轴

# Canny 边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')  # 使用灰度色图显示边缘图
plt.title("Canny 边缘检测")
plt.axis('off')  # 不显示坐标轴

# 显示结果
plt.tight_layout()
plt.show()
```
