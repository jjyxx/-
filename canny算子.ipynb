```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
src = cv2.imread("D:/picture/school.png")
if src is None:
    print("Could not load image...")
    exit()

# 
gray_src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

#  Canny 
# Canny 
low_threshold = 100
high_threshold = 200
edges = cv2.Canny(gray_src, low_threshold, high_threshold)

#  plt 
plt.figure(figsize=(10, 8))
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False    # 

# 
plt.subplot(1, 2, 1)
#plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))  #  RGB  matplotlib
plt.imshow(gray_src,cmap='gray')
plt.title("")
plt.axis('off')  # 

# Canny 
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')  # 
plt.title("Canny ")
plt.axis('off')  # 

# 
plt.tight_layout()
plt.show()
```
