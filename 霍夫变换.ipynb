```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
src = cv2.imread("D:/picture/school.png")
if src is None:
    print("Could not load image...")
    exit()

# 
gray_src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

#  Canny 
dst = cv2.Canny(gray_src, 100, 200, 3)

# 
plines = cv2.HoughLinesP(dst, 1, np.pi / 180, 70, minLineLength=100, maxLineGap=50)

# 
if plines is not None:
    for line in plines:
        x1, y1, x2, y2 = line[0]
        cv2.line(src, (x1, y1), (x2, y2), (255, 255, 0), 2, cv2.LINE_AA)

#  plt 
plt.figure(figsize=(12, 6))
plt.rcParams['font.sans-serif'] = ['SimHei']  # 
plt.rcParams['axes.unicode_minus'] = False    # 
# 
plt.subplot(1, 2, 1)
#plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))  #  RGB  matplotlib
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
plt.title("")
plt.axis('off')  # 

# 
plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.title("Canny ")
plt.axis('off')  # 

# 
plt.tight_layout()
plt.show()
```
