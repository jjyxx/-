```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
src = cv2.imread("D:/picture/school.png")
if src is None:
    print("Could not load image...")
    exit()

# 转换为灰度图像
gray_src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# 使用 Canny 边缘检测
dst = cv2.Canny(gray_src, 100, 200, 3)

# 霍夫变换进行直线检测
plines = cv2.HoughLinesP(dst, 1, np.pi / 180, 70, minLineLength=100, maxLineGap=50)

# 标记检测到的直线
if plines is not None:
    for line in plines:
        x1, y1, x2, y2 = line[0]
        cv2.line(src, (x1, y1), (x2, y2), (255, 255, 0), 2, cv2.LINE_AA)

# 使用 plt 显示图像
plt.figure(figsize=(12, 6))
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
# 原始图像
plt.subplot(1, 2, 1)
#plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))  # 转换为 RGB 格式以适应 matplotlib
plt.imshow(cv2.cvtColor(src, cv2.COLOR_BGR2RGB))
plt.title("检测到的直线")
plt.axis('off')  # 不显示坐标轴

# 边缘检测结果
plt.subplot(1, 2, 2)
plt.imshow(dst, cmap='gray')
plt.title("Canny 边缘检测")
plt.axis('off')  # 不显示坐标轴

# 显示结果
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
src = cv2.imread("D:/picture/school.png")
if src is None:
    print("无法加载图像...")
    exit()

# 转换为灰度图像
gray_src = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# 使用 Canny 边缘检测
dst = cv2.Canny(gray_src, 100, 200, 3)

# 霍夫变换进行直线检测
plines = cv2.HoughLinesP(dst, 1, np.pi / 180, 70, minLineLength=100, maxLineGap=50)

# 在灰度图像上绘制检测到的直线
if plines is not None:
    for line in plines:
        x1, y1, x2, y2 = line[0]
        cv2.line(gray_src, (x1, y1), (x2, y2), (255), 2, cv2.LINE_AA)  # 绘制白色的直线

# 使用 matplotlib 显示图像
plt.figure(figsize=(8, 8))

# 显示带直线的灰度图像
plt.imshow(gray_src, cmap='gray')
plt.title("灰度图像与检测到的直线")
plt.axis('off')  # 关闭坐标轴
plt.show()
```
