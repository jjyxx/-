```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# 读取图像
img = cv.imread("D:/picture/school.png", 0)

# 检查图像是否加载成功
if img is None:
    print("图像加载失败，请检查文件路径是否正确。")
else:
    # 灰度级切片参数
    lower_bound = 100
    upper_bound = 200
    high_value = 255

    # 创建切片后的图像
    sliced_img = np.zeros_like(img)
    sliced_img[(img >= lower_bound) & (img <= upper_bound)] = high_value

    # 设置中文字体和负号显示
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
    plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

    # 显示原始图像和切片后的图像
    plt.figure(figsize=(12, 10))

    # 显示原始图像
    plt.subplot(121)
    plt.imshow(img, cmap='gray')
    plt.title('原始图像')
    plt.axis('off')

    # 显示灰度级切片图像
    plt.subplot(122)
    plt.imshow(sliced_img, cmap='gray')
    plt.title('灰度级切片图像')
    plt.axis('off')

    # 定义灰度值范围
    r_values = np.arange(0, 256)
    T_values = np.zeros_like(r_values)
    T_values[(r_values >= lower_bound) & (r_values <= upper_bound)] = high_value

    # 绘制灰度级切片的转换函数 T(r)
    plt.figure(figsize=(6, 4))
    plt.plot(r_values, T_values, color='blue')
    plt.xlabel('r')
    plt.ylabel('s')
    plt.grid(True)
    plt.title('灰度级切片转换函数 T(r)')
    plt.show()
```

```python

import cv2 as cv
import matplotlib.pyplot as plt

# 读取图像
img = cv.imread("D:/picture/school.png", 0)

# 检查图像是否加载成功
if img is None:
    print("图像加载失败，请检查文件路径是否正确。")
else:
    # 灰度级切片参数
    lower_bound = 100
    upper_bound = 200
    high_value = 255

    # 创建切片后的图像
    sliced_img = np.zeros_like(img)
    sliced_img = img.copy()
    sliced_img[(img >= lower_bound) & (img <= upper_bound)] = high_value
    
    # 设置中文字体和负号显示
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
    plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

    # 显示原始图像和切片后的图像
    plt.figure(figsize=(12, 10))

    # 显示原始图像
    plt.subplot(121)
    plt.imshow(img, cmap='gray')
    plt.title('原始图像')
    plt.axis('off')

    # 显示灰度级切片图像
    plt.subplot(122)
    plt.imshow(sliced_img, cmap='gray')
    plt.title('灰度级切片图像')
    plt.axis('off')

    # 定义灰度值范围
    r_values = np.arange(0, 256)
    T_values = np.zeros_like(r_values)
    r_values = np.arange(0, 256)
    T_values = r_values.copy()  # 让 T_values 初始值和 r_values 相同

    # 将灰度值在指定范围内的像素设置为高亮值
    T_values[(r_values >= lower_bound) & (r_values <= upper_bound)] = high_value


    # 绘制灰度级切片的转换函数 T(r)
    plt.figure(figsize=(6, 4))
    plt.plot(r_values, T_values, color='blue')
    plt.xlabel('r')
    plt.ylabel('s')
    plt.grid(True)
    plt.title('灰度级切片转换函数 T(r)')
    plt.show()
```
