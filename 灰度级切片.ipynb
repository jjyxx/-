```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# 
img = cv.imread("D:/picture/school.png", 0)

# 
if img is None:
    print("")
else:
    # 
    lower_bound = 100
    upper_bound = 200
    high_value = 255

    # 
    sliced_img = np.zeros_like(img)
    sliced_img[(img >= lower_bound) & (img <= upper_bound)] = high_value

    # 
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 
    plt.rcParams['axes.unicode_minus'] = False    # 

    # 
    plt.figure(figsize=(12, 10))

    # 
    plt.subplot(121)
    plt.imshow(img, cmap='gray')
    plt.title('')
    plt.axis('off')

    # 
    plt.subplot(122)
    plt.imshow(sliced_img, cmap='gray')
    plt.title('')
    plt.axis('off')

    # 
    r_values = np.arange(0, 256)
    T_values = np.zeros_like(r_values)
    T_values[(r_values >= lower_bound) & (r_values <= upper_bound)] = high_value

    #  T(r)
    plt.figure(figsize=(6, 4))
    plt.plot(r_values, T_values, color='blue')
    plt.xlabel('r')
    plt.ylabel('s')
    plt.grid(True)
    plt.title(' T(r)')
    plt.show()
```
