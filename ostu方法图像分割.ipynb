```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像（灰度模式）
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 计算图像的梯度
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# 计算梯度幅度
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# 显示结果
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False



# 指定一个阈值 T（例如，梯度幅度的 90% 分位数）
T = np.percentile(gradient_magnitude, 90)
_, gT = cv2.threshold(gradient_magnitude, T, 255, cv2.THRESH_BINARY)



# 使用阈值 T 进行全局分割
fixed_threshold = T  # 你可以调整这个值
_, segmented_image = cv2.threshold(image, fixed_threshold, 255, cv2.THRESH_BINARY)




plt.figure(figsize=(12, 8))
# 显示梯度幅度图像
plt.subplot(131)
plt.title('梯度幅度图像')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

# 显示二值边缘图像 gT(x, y)
plt.subplot(132)
plt.title('二值边缘图像 gT(x, y)')
plt.imshow(gT, cmap='gray')
plt.axis('off')


# 显示分割结果
plt.subplot(133)
plt.imshow(segmented_image, cmap='gray')
plt.title(f'使用固定阈值 T ({fixed_threshold}) 进行全局分割')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像（请确保使用灰度模式读取图像）
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 显示结果
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


# 使用 Otsu 方法自动计算最佳阈值并进行图像分割
# 设置阈值为 0，并使用 THRESH_OTSU 标志来启用 Otsu 方法
ret, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# 显示 Otsu 分割后的图像
plt.imshow(otsu_thresh, cmap='gray')
plt.title('Otsu方法进行全局分割')
plt.axis('off')
plt.show()

# 输出计算出的最佳阈值
print("Otsu's threshold: ", ret)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像（灰度模式）
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 计算图像的梯度
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# 计算梯度幅度
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# 显示结果
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False



# 指定一个阈值 T（例如，梯度幅度的 90% 分位数）
T = np.percentile(gradient_magnitude, 90)
_, gT = cv2.threshold(gradient_magnitude, T, 255, cv2.THRESH_BINARY)



# 使用阈值 T 进行全局分割
fixed_threshold = T  # 你可以调整这个值
_, segmented_image = cv2.threshold(image, fixed_threshold, 255, cv2.THRESH_BINARY)

# 使用 Otsu 方法自动计算最佳阈值并进行图像分割
# 设置阈值为 0，并使用 THRESH_OTSU 标志来启用 Otsu 方法
ret, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)



# 输出计算出的最佳阈值
print("Otsu's threshold: ", ret)


plt.figure(figsize=(12, 8))
# 显示分割结果
plt.subplot(121)
plt.imshow(segmented_image, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title(f'使用固定阈值 T ({fixed_threshold}) 进行全局分割')
#plt.axis('off')

# 显示 Otsu 分割后的图像
plt.subplot(122)
plt.imshow(otsu_thresh, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Otsu方法进行全局分割')

plt.show()
```
