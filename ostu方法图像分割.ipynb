```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# 
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False



#  T 90% 
T = np.percentile(gradient_magnitude, 90)
_, gT = cv2.threshold(gradient_magnitude, T, 255, cv2.THRESH_BINARY)



#  T 
fixed_threshold = T  # 
_, segmented_image = cv2.threshold(image, fixed_threshold, 255, cv2.THRESH_BINARY)




plt.figure(figsize=(12, 8))
# 
plt.subplot(131)
plt.title('')
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')

#  gT(x, y)
plt.subplot(132)
plt.title(' gT(x, y)')
plt.imshow(gT, cmap='gray')
plt.axis('off')


# 
plt.subplot(133)
plt.imshow(segmented_image, cmap='gray')
plt.title(f' T ({fixed_threshold}) ')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
image = cv2.imread('D:\picture\school.png', cv2.IMREAD_GRAYSCALE)

# 
gradient_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# 
gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False



#  T 90% 
T = np.percentile(gradient_magnitude, 90)
_, gT = cv2.threshold(gradient_magnitude, T, 255, cv2.THRESH_BINARY)



#  T 
fixed_threshold = T  # 
_, segmented_image = cv2.threshold(image, fixed_threshold, 255, cv2.THRESH_BINARY)

#  Otsu 
#  0 THRESH_OTSU  Otsu 
ret, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)



# 
print("Otsu's threshold: ", ret)


plt.figure(figsize=(12, 8))
# 
plt.subplot(121)
plt.imshow(segmented_image, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title(f' T ({fixed_threshold}) ')
#plt.axis('off')

#  Otsu 
plt.subplot(122)
plt.imshow(otsu_thresh, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Otsu')

plt.show()
```
