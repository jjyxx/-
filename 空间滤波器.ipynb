```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def junzhi(img):

# 应用均值滤波器，使用3x3的滤波器窗口
    return cv.blur(img, (15, 15))

def fangkuang(img):
    # 应用方框滤波器，使用3x3的滤波器窗口
    return cv.boxFilter(img,ddepth=-1, ksize=(15, 15),)
    
def gaosi(img):
    # 应用高斯滤波器，使用3x3的滤波器窗口
    return cv.GaussianBlur(img, (15, 15), 0)
    
```

```python
# 加载图像并转换为灰度图像
#img = cv.imread("D:/picture/school.png")
img= cv.imread("D:/picture/school.png", cv.IMREAD_GRAYSCALE)
mean_img=junzhi(img)
box_img=fangkuang(img)
gaosi_img=gaosi(img)
#show_four_pictures(img, mean_img,box_img,gaosi_img)
# 将图像从BGR格式转换为RGB格式，以便用matplotlib显示
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
mean_rgb = cv.cvtColor(mean_img, cv.COLOR_BGR2RGB)
box_rgb = cv.cvtColor(box_img, cv.COLOR_BGR2RGB)
gaussian_rgb = cv.cvtColor(gaosi_img, cv.COLOR_BGR2RGB)



plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
# 设置图形窗口的宽度为12英寸，高度为8英寸
plt.figure(figsize=(12,8 ))
plt.subplot(4, 2, 1)
plt.imshow(img_rgb,cmap='gray')
plt.title('原始图像')
plt.axis('off')

plt.subplot(4, 2, 2)
plt.imshow(mean_rgb,cmap='gray')
plt.title('均值滤波处理后的图像')
plt.axis('off')

plt.subplot(4, 2, 3)
plt.imshow(img_rgb,cmap='gray')
plt.title('原始图像')
plt.axis('off')

plt.subplot(4, 2, 4)
plt.imshow(box_rgb,cmap='gray')
plt.title('方框滤波处理后的图像')
plt.axis('off')

plt.subplot(4, 2, 5)
plt.imshow(img_rgb,cmap='gray')
plt.title('原始图像')
plt.axis('off')


plt.subplot(4, 2, 6)
plt.imshow(gaussian_rgb,cmap='gray')
plt.title('高斯滤波处理后的图像')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 加载图像并转换为灰度图像
img = cv.resize(cv.imread("D:/picture/school.png"), (400, 200))
mean_img=junzhi(img)
box_img=fangkuang(img)
gaosi_img=gaosi(img)
#show_four_pictures(img, mean_img,box_img,gaosi_img)
# 将图像从BGR格式转换为RGB格式，以便用matplotlib显示
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
mean_rgb = cv.cvtColor(mean_img, cv.COLOR_BGR2RGB)
box_rgb = cv.cvtColor(box_img, cv.COLOR_BGR2RGB)
gaussian_rgb = cv.cvtColor(gaosi_img, cv.COLOR_BGR2RGB)

# 设置图形窗口的宽度为12英寸，高度为8英寸
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(mean_rgb)
plt.title('Mean Filter')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(box_rgb)
plt.title('Box Filter')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(gaussian_rgb)
plt.title('Gaussian Filter')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
均值滤波器:图像较为模糊，去除了部分噪声，但细节部分也随之丢失，尤其是在细腻的边缘区域。

方框滤波器（Box Filter）：与均值滤波器类似，但是高亮区域的边缘模糊得更明显，细节损失较多。

高斯滤波器（Gaussian Filter）：平滑效果较好，同时保留了更多的细节。高斯滤波器对细节的处理比前两者更加平衡，能够有效去除噪声的同时，保持轮廓的清晰度。
```

```python
在使用 OpenCV 处理图像时，图像默认以 BGR（蓝-绿-红）格式存储和显示。相反，matplotlib 使用的是 RGB（红-绿-蓝）格式。
因此，当你在 OpenCV 中读取图像后，如果直接用 matplotlib 显示，它的颜色会出现不正常的现象，比如蓝色和红色会被互换。
OpenCV 本身提供了显示图像的功能，使用 cv2.imshow() 可以直接在一个窗口中显示图像。
```
