```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
src = cv2.imread("D:/picture/school.png")

# 将图像转换为灰度图像
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

# 初始化 HOG 描述符
hog = cv2.HOGDescriptor()

# 计算图像的 HOG 特征
hog_features = hog.compute(gray)
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体显示中文
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
# 归一化 HOG 特征
norm_hog_features = cv2.normalize(hog_features, None, 0, 255, cv2.NORM_MINMAX)

# 可视化归一化后的HOG直方图（默认每个特征向量的大小为36, 这代表每个cell的梯度方向）
plt.figure(figsize=(8, 5))
plt.hist(norm_hog_features.ravel(), bins=50, range=(0, 255))
plt.xlabel("HOG特征值")
plt.ylabel("频率")
plt.grid(True)
plt.show()
```
